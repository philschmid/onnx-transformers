```python
import os
from dataclasses import dataclass
from onnxruntime.transformers.convert_generation import main
from onnxruntime.transformers.benchmark_helper import Precision


@dataclass
class OptimizationParameter:
    model_name_or_path: str
    model_type: str = "gpt2"  # "t5", "gpt2", "mt5"
    output: str = "t5_beam.onnx"
    precision: Precision = Precision.FLOAT32
    use_gpu: bool = True
    min_length: int = 1
    max_length: int = 20
    num_beams: int = 4
    num_return_sequences: int = 1
    length_penalty: float = 1.0
    repetition_penalty: float = 1.0
    vocab_size: int = -1
    cache_dir: str = os.path.join(".", "cache_models")
    decoder_onnx: str = ""
    encoder_decoder_init_onnx: str = ""
    use_external_data_format: bool = False
    run_shape_inference: bool = False
    disable_shared_initializers: bool = False
    output_sequences_scores: bool = False
    output_token_scores: bool = False
    early_stopping: bool = False
    no_repeat_ngram_size: int = 0
    vocab_mask: bool = False
    prefix_vocab_mask: bool = False
    custom_attention_mask: bool = False
    disable_parity: bool = False
    torch_performance: bool = False
    total_runs: int = 1
    save_test_data: bool = False
    verbose: bool = False


_t5 = OptimizationParameter(
    model_name_or_path="t5-small",
    model_type="t5",
)
```

```python
# can be fixed by forking onnxruntime and fixing their path structure
main(["--model_name_or_path", _t5.model_name_or_path,"--model_type", _t5.model_type, "--use_gpu", "--torch_performance", "--vocab_mask", "--output", "./models/t5/onnx_models/t5_small_beam_search.onnx"])
```
